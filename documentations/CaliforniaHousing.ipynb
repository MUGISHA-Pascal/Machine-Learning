# California housing dataset from Sklearn.datasets with linear Regression

# importing the dataset

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

# loading the data using pandas into y and x axis

```python
data = fetch_california_housing()
pdData  = pd.DataFrame(data.data,columns=data.feature_names)
pdData['MedHouseVal']=data.target

x = pdData.drop('MedHouseVal',axis=1)
y = pdData['MedHouseVal']
```

# splitting the data into test and train portions for essential training of the model

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

# using standard scaler for preprocessing

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

# loading the model algorithms (linear regression)

```python
model = LinearRegression()
model.fit(x_train_scaled,y_train)
```

# predicting the accuracy using r2_score

```python
y_pred = model.predict(x_test_scaled)
accuracy = r2_score(y_test,y_pred)
print(f"accuracy is {accuracy}")
```

# using matplotlib to plot the 2d cartesian plane

```python
plt.scatter(y_test,y_pred,color="blue")
plt.xlabel("test values")
plt.ylabel("prediction")
plt.title("the california housing dataset 2d cartesian plane")
plt.show()
```
